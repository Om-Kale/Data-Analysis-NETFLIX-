# Exploratory Data Analysis

# country

```python
# MAnupulating the data of countries 
ax = sns.countplot(data = df,x ='Country')
sns.set(rc= {'figure.figsize':(19,5)})
```

```python
# Country vs Monthly Revenue 
sales_C = df.groupby(['Country'], as_index = False)['Monthly Revenue'].sum().sort_values(by = 'Monthly Revenue',ascending = False)
sns.set(rc = {'figure.figsize':(15,6)})
sns.barplot(data = sales_C ,x = 'Country',y = 'Monthly Revenue')
```

we clearly saw that NETFLIX users mostly acitve in United States, Spain and Canada

# Age

```python
ax= sns.countplot(data = df, x = 'Age',hue = 'Gender')
```

```python
ax= sns.countplot(data = df, x = 'Age',hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```
